````python
df.distance[5514988]
````

````python
#Since distance cannot be negative we will be droping the negative values
df.drop(df.index[df.distance <= 0])
````

````python
#dropping the null values
df = df.dropna(axis = 0)
````

````python
#dropping the unneccesarry data which is not helpful in analysis hence decreasing the dimension of the dataframe

df = df.drop(['start_lat','start_lng','end_lat','end_lng'], axis = 1)
````

````python
df
````

````python
df.isnull().sum()
````

````python
#As the day_week_start contains 1.0, 2.0 as a numeric value it would be better to have Sunday, Monday... for better understanding
df = df.replace({1.0: 'Sunday', 2.0: 'Monday', 3.0: 'Tuesday', 4.0: 'Wednesday', 5.0: 'Thursday', 6.0: 'Friday', 7.0: 'Saturday'})
````

````python
#drop the duplicate values if present in the dataframe
df.drop_duplicates()
df
````

````python
import datetime
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
````

````python
df.dtypes
````

````python
# Calculating the time taken by each customer for easier calculation
df['time_taken'] = df['ended_at'] - df['started_at']
df.time_taken
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
from datetime import date
df_casual['started_at'] = pd.to_datetime(df_casual['started_at']).dt.normalize()
df_casual['ended_at'] = pd.to_datetime(df_casual['ended_at']).dt.normalize()
````

````python
df_date_casual = pd.Series(df_casual['started_at']).value_counts().sort_index()
df_date_casual.index = pd.DatetimeIndex(df_date_casual.index)
df_date_casual = df_date_casual.rename_axis('date').reset_index(name='counts')
df_date_casual
````
